# Nova seção

```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas ["Nome_Departamento"] = vendas ["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
vendas[vendas["estado"].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas["estado"].fillna("MS", inplace = True)
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas.info()
```

```python
media_preco = vendas ["Preço"].mean()
vendas["Preço"].fillna(media_preco, inplace = True)
```

```python
media_preco
```

```python
vendas.info()
```

```python
mediana_preco = vendas["Preço"].median()
mediana_preco
```

```python
vendas[["Preço", "Preço_com_frete"]].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")
preco_correto = vendas.query("Preço < Preço_com_frete")
```

```python
preco_correto.describe()
```

```python
preco_correto.head()
```

```python
preco_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
preco_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
qtd_data = preco_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False)
qtd_data.head()
```

```python
preco_correto["Mês"] = preco_correto["Data"].dt.month
preco_correto.head()
```

```python
 preco_correto.groupby("Mês").idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel("/content/cliente_varejo.xlsx")
cliente.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_cliente = preco_correto.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(7,5))

plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")

plt.ylabel("Média de idade")


plt.title("Idade média por Bandeira", loc="center", fontsize=18, fontweight="bold", color="black");
```

```python
fig = plt.figure(figsize=(7,5))

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")

plt.ylabel("Renda média")

plt.title("Renda média por canal de venda", loc="center", fontsize=18, fontweight="bold", color="black");
```

```python
venda_por_data = preco_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(18,10))


plt.xlabel("Data da venda")
plt.ylabel("Quantidade de vendas")
x = venda_por_data["Data"]
y = venda_por_data["idcompra"]

plt.plot(x,y, color = "green", linewidth = "2")
plt.show()
```

```python
agg_dept_preco = round(preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(60,25))

plt.bar(agg_dept_preco["Nome_Departamento"], agg_dept_preco["Preço_com_frete"], color="green")

plt.ylabel("Preço médio com frete")

plt.title("Preço médio por departamento", loc="center", fontsize=18, fontweight="bold", color="black");
```
